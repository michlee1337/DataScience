```r
install.packages("Matching")
library(Matching)

data(lalonde)

x <- lalonde$treat  # binary variable (0/1), random assignment to training
y <- lalonde$re78   # real earnings in 1978, after the training

plot(x, y,          # what follows is a bunch of commands for a pretty plot
     pch = 19, col = "blue", cex = 0.5, 
     main = "Lalonde -- An RCT to Estimate Impact of Training on Earnings", 
     xlab =  "Assignment to Training: \n(0 means \"control group\" or 
              \"not-assigned\"---1 means \"treatment group\" or
  \"assigned\")", ylab = "$ earned in 1978",
     ylim = c(0, 30000),
     xaxt = "n")    # this line tells R **DO NOT** draw the x-axis


axis(side = 1, at = c(0,1)) # draws x axis and sets 2 tick marks at 0 and 1

regression_1 <- lm(re78 ~ treat, data = lalonde)
summary(regression_1)

abline(regression_1, col = "red", lwd = 3)   # draws the regression line
```

```r
lm(re78 ~ treat, data = lalonde)
```

```r
treat <- lalonde[lalonde$treat==1,]
no_treat <- lalonde[lalonde$treat==0,]
head(treat)
```
