### Understanding Factors

```r
factor_vars <- c()

for (i in seq(1:ncol(foo))) {
  if(class(foo[,i]) == "factor"){
      factor_vars <- c(factor_vars, names(foo[i]))
  }
}
```

```r
percentage <- prop.table(table(foo$Status)) * 100
tab <- cbind(freq=table(foo$Status), percentage=percentage)
tab
```

```r
# print frequency tables for all factor variables

freq_tables <- c()
for (col in factor_vars){
    table_name <- paste("table", col, sep = "")
    percentage <- prop.table(table(foo[col])) * 100
    assign(table_name, cbind(freq=table(foo[col]), percentage=percentage))
}

for (col in factor_vars){
    print(get(paste("table", col, sep = "")))
    print('____________')
}
```
