```python
import numpy as np
import math
from matplotlib import pyplot as plt

def diff(x,y):
    return(((1-y)*y) - 0.1)

x = np.linspace(0,0.5,10)
y = np.linspace(0.5,1.5,11)

for j in x:
    for k in y:
        slope = diff(j,k)
        domain = np.linspace(j-0.07,j+0.07,2)
        def fun(x1,y1):
            z = slope*(domain-x1)+y1
            return z
        plt.plot(domain,fun(j,k),solid_capstyle='projecting',solid_joinstyle='bevel')

plt.title("Slope field y'")
plt.grid(True)
plt.show()
```
