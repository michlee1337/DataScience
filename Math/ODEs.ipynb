```python
import numpy as np
import math
from matplotlib import pyplot as plt

def diff(x,y):
    return(2*y-y**3)

x = np.linspace(-5,5,10)
y = np.linspace(-3,3,11)

for j in x:
    for k in y:
        slope = diff(j,k)
        domain = np.linspace(j-0.07,j+0.07,2)
        def fun(x1,y1):
            z = slope*(domain-x1)+y1
            return z
        plt.plot(domain,fun(j,k),solid_capstyle='projecting',solid_joinstyle='bevel')

plt.title("Slope field y'")
plt.grid(True)
plt.show()
```

```python
def bisection(f,start,end):
    mid = start + (end-start)/2
    print(mid, f(mid))
    
    if abs(f(mid) - 0) < 0.05:
        return(mid)
    elif f(mid) < 0:
        return(bisection(f,mid,end))
    else:
        return(bisection(f,start,mid))
```
