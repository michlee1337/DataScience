Most general solution for a system $ax''+bx'+cx = 0$ is $$c_1e^{\lambda_1}+c_2e^{\lambda_2}$$ where $$\lambda_1/\lambda_2 = \frac{-b\pm \sqrt{b^2-4ac}}{2a}$$

Substituting in the values from the ODE:

$$c_1e^{\lambda_1t}+c_2e^{\lambda_2t}$$ where $$\lambda_1/\lambda_2 = \frac{\pm \sqrt{-4\omega_0^2}}{2}$$
$$\lambda_1/\lambda_2 = \frac{\pm 2i\omega_0}{2}$$
$$\lambda_1/\lambda_2 = \pm i\omega_0$$


Thus the most general solution for $x(t)$ will be:

$$c_1e^{i \omega_0 t}+c_2e^{-i \omega_0 t}$$
$$c_1(cos(\omega_0t)+isin(\omega_0t)) + c_2(cos(\omega_0t-isin(\omega_0t))$$
$$(c_1+c_2)cos(\omega_0t)+i(c_1-c_2)sin(\omega_0t)$$


**Solving IVP**

$$(c_1+c_2)cos(2(0))+i(c_1-c_2)sin(2(0)) = 0$$
$$c_1+c_2 = 1$$

$$-2(c_1+c_2)sin(2(0))+i2(c_1-c_2)cos(2(0)) = 0$$
$$i2(c_1-c_2) = 0$$

$$c_1-c_2 = 0$$

Thus, we have the two equations:
$$c_1+c_2 = 1$$
$$c_1-c_2 = 0$$

Rearranging this, we get
$$c_1 = c_2 = 0.5$$

Thus,
$$x(t) = cos(2t)$$

$$x_p' = -Asin(t)+ Bcos(t)$$
$$x_p'' = -Acos(t)-Bsin(t)$$

Subbing these values back into the original ODE:
$$-Acos(t)-Bsin(t) + \omega_0^2(Acos(t)+Bsin(t)) = q(x)$$
$$(\omega_0^2A-A)cos(t)+(\omega_0^2B-B)sin(t)) = q(x)$$

Now that both the left and the right are expressed in terms of periodic functions, we can solve for the coefficients

## a)

$$V = \frac{1}{\pi}\int_{-\pi}^\pi f(t)g(t) dt$$ for all $f(t)$ and $g(t)$


**Closed under Addition**  
$$<f(t),g(t)+h(t)> = \frac{1}{\pi}\int_{-\pi}^\pi (f(t))(g(t)+h(t)) dt$$ 
$$= \frac{1}{\pi}\int_{-\pi}^\pi f(t)g(t)+f(t)h(t) dt$$ 
$$= \frac{1}{\pi}\int_{-\pi}^\pi f(t)g(t) dt + \frac{1}{\pi}\int_{-\pi}^\pi f(t)h(t) dt$$ 
$$<f(t),g(t)+h(t)> = <f(t),g(t))> + <f(t),h(t))>$$ 


**Closed under Scalar Multiplication**    
$$<f(t),ug(t)> = \frac{1}{\pi}\int_{-\pi}^\pi uf(t)g(t) dt$$ 
Because $u$ is a scalar not dependent on $t$
$$<f(t),ug(t)> = u\frac{1}{\pi}\int_{-\pi}^\pi f(t)g(t) dt$$ 
$$<f(t),ug(t)> = u<f(t),g(t)>$$ 


**Contains Zero**   
In this context, the zero vector is a function that equals to zero
$$<f(t),0> = \frac{1}{\pi}\int_{-\pi}^\pi f(t)(0) dt$$ 
$$<f(t),0> = 0$$ 


## b)

Orthogonal basis means that the inner product of any two vectors in the space is zero.

We know that $sin(at)sin(bt) = 0$,$cos(at)cos(bt) = 0$, and $sin(t)cos(t) = 0$ when $a != b$

Thus the ortogonal basis is $\{sin(ax),cos(ax)\}$ for every possible unique value of $a$. 

## c)

Orthonormal basis has the additional condition that the norm of every vector is 1.

Given the norm definition that $$|f(t)| = <f(t),f(t)>$$

$$<f(t),f(t)> = \frac{1}{\pi}\int_{-\pi}^\pi f(t)^2 dt$$ 
Given that we know that the integral of any periodic function squared is $\pi$, in order to make their inner product 1, we must divide every vector by $\pi$.

Thus the orthonormal basis of the inner space of periodic functions is just $\{\frac{sin(ax)}{\pi},\frac{cos(ax)}{\pi}\}$ for every possible unique value of $a$.
