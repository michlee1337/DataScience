```python
import numpy as np
import matplotlib.pyplot as plt
L = 1
tmax = 1
nx = 11
nt= 11

def HeatSolver(L, tmax, nx, nt):
    dx = L/(nx-1)
    dt = tmax/(nt-1)
    r = dt/(dx**2)
    
    u= np.zeros((nt, nx))
    u[0,:] = [np.sin(np.pi*x/L) for x in np.linspace(0,L,nx)]
    
    for t in range(1,nt):
        for x in range(1,nx-1): 
            u[t,x] = u[t-1,x] + r*(u[t-1,x+1]-1*u[t-1,x]+2*u[t-1,x-1])
    return u
```

```python
result = HeatSolver(L, tmax, nx, nt)
```

```python
x = np.linspace(0,L,nx)
t = np.linspace(0,tmax,nt)
X,T = np.meshgrid(x,t)
plt.contour(X,T,result);
```
