Describes how extracurricular engagement (percentile of student on attendence of optional school activities) is affected by both student diligence and family income. While student diligence could increase the amount of optional school activities attended, family income is very likely a limiting factor. students with more financial restrictions would have priorities above optional school activities.

Given the dilligence and family income of students, predict their extracurricular engagment level

```r
dilligence <- rnorm(200,1,10) 
family_income <- rnbinom(200,10,.5)

academic_engagement <- 50 + 10*dilligence + (500*family_income) + rnorm(200,0,1000) 
```

```r
hist(dilligence)
```

```r
hist(family_income)
```

```r
hist(academic_engagement)
```

```r
# see results
lm(academic_engagement ~ dilligence + family_income) 
```
