# Generate Data

```r
x = rnorm(999,0,10)
y = 2*x + rnorm(999,0,10)
data = data.frame(x,y)

plot(data$x,data$y)
fit1 = lm(y ~ x)
abline(fit1)
```

```r
data = rbind(data,c(3000,-400))
plot(data$x,data$y)
fit2 = lm(data$y ~ data$x)
abline(fit2)
```

summary(fit1)

```r
summary(fit2)
```
