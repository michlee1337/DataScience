```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
x = np.linspace(-10,10,200)
```

```python
g1 = plt.plot(x, scipy.stats.gamma.pdf(x,3))
g2 = plt.plot(x, scipy.stats.gamma.pdf(x,5))

g1+g2
```

```python
plt.plot(x, scipy.stats.gamma.pdf(x,3))
```

```python
gs = np.random.gamma(3,1,500)

plt.hist(gs,density=True)
plt.plot(x, scipy.stats.gamma.pdf(x,3))
plt.show()
```

```python
n1 = plt.plot(x, scipy.stats.norm.pdf(x,3))
n2 = plt.plot(x, scipy.stats.norm.pdf(x,5))

n1+n2
```

```python
ns = np.random.normal(3,1,500)

plt.hist(ns,density=True)
plt.plot(x, scipy.stats.norm.pdf(x,3))

plt.show()
```

```python
b1 = plt.plot(x, scipy.stats.beta.pdf(x,2,3))
b2 = plt.plot(x, scipy.stats.beta.pdf(x,6,5))

b1+b2
```

```python
bs = np.random.beta(2,3,500)

plt.hist(bs,density=True)
plt.plot(x, scipy.stats.beta.pdf(x,2,3))
plt.show()
```
