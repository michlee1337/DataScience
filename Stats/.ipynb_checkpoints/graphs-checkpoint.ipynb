```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edges_from(
    [('x_0', 'x_1'), ('v_0', 'x_1'),('sd_x', 'x_1')])

val_map = {'x_1': 0,
           'x_0': 1,
           'v_0': 1,
           'sd_x': 1}

values = [val_map.get(node, 0.25) for node in G.nodes()]

nx.draw(G, cmap = plt.get_cmap('jet'), node_color = values)
plt.show()
```

```python
G = nx.Graph()
G.add_node("x_0")
G.add_node("v_0")
G.add_node("x_0", name="x_0")

nx.draw(G, cmap = plt.get_cmap('jet'),node_color='y', node_size=2000)

# labels
labels={}
labels["x_0"]="$x_0$"
#labels["v_0"]="$v_0$"

pos=nx.spring_layout(G) # positions for all nodes
nx.draw_networkx_labels(G,pos,labels,font_size=16)
plt.show()
```
