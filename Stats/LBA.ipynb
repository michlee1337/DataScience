```python
import pandas as pd
```

```python
# get columns and rows
data = pd.read_csv("LBA.csv", header=1)
data = data.drop(0)
```

```python
# deal w grocery name
def firstUpper(s):
    # edge case: deal w tesco exp cause its diff
    if s.lower() == "tesco express":
        return("TESCOEXP")
    return(s.split(" ")[0].upper())
data["Grocery store"] = data["Grocery store"].apply(firstUpper)
```
