```python
import scipy.stats
import matplotlib.pyplot as plt 
import numpy as np 
import math
```

```python
# mixture of Gaussians: number of minervans in a city 
# support: real numbers
# skewed: bimodal with one heavier than the other
# probability mass mostly at 0
# parameters: mean and sd of each of the 2 modals and the 2 relative weights of the two modals


def myPdf(x, m1,sd1,pi1,m2,sd2,pi2):
    term1 = (pi1/(math.sqrt(2*math.pi*sd1**2)))*math.e**(-(x-m1)**2/(2*sd1**2))
    term2 = (pi2/(math.sqrt(2*math.pi*sd2**2)))*math.e**(-(x-m2)**2/(2*sd2**2))
    return(term1+term2)

def myCdf(x, m1,sd1,pi1,m2,sd2,pi2):
    Z1 = (x-m1)/sd1
    Z2 = (x-m2)/sd2
    return(scipy.stats.norm.cdf(Z1+Z2))

def myPpf(x, m1,sd1,pi1,m2,sd2,pi2):
    return([scipy.stats.norm.ppf(i) for i in myCdf(x, m1,sd1,pi1,m2,sd2,pi2)])
    
    
```

```python
x = np.linspace(1, 250,1000) 
plt.plot(x, myPdf(x, 1,10,0.6,200,10,0.4))
```

```python
plt.plot(x, myCdf(x, 1,10,0.6,200,10,0.4))
```

```python
plt.plot(x, myCdf(x, 1,10,0.6,200,10,0.4))
```

```python
plt.plot(x, myPpf(x, 1,10,0.6,200,10,0.4))
```
