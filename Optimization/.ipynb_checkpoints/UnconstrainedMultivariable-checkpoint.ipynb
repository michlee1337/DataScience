```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import multiprocessing
import time

def describeCrit2(f,x,y):
    '''
    find location and type of all critical points
    for a function f(x,y)

    input: function
    output: [(critical point, type), (critical point, type), ...]
    '''
    # find critical points
    critPoints = critCoords2(f,x,y)
    
    # find type of critical point (2nd erivative testd)
    hessian = hessian2(f,x,y)
    types = typeCrit2(hessian, critPoints)
    
    res = [(critPoints[i],types[i]) for i,_ in enumerate(critPoints)]
    return(res)
    

def critCoords2(f,x,y):
    '''
    find coordinates all critical points
    for a function f(x,y)

    input: function
    output: [(x1, y1), (x2, y2), ...]
    '''
    
    # find the partials
    px = f.diff(x)
    py = f.diff(y)
    
    # express y in terms of x
    y_expX = solveset(py,y).args[0]
    
    # substitute to solve for x
    x_eq = px.subs(y,y_expX)
    
    # solve for x coordinates of critical points
    x_coords = [i for i in solveset(x_eq,x).args]
    
    # solve for y coordinates of critical points
    y_coords = [y_expX.subs(x,val) for val in x_coords]
    return([(x_coords[i], y_coords[i]) for i,_ in enumerate(x_coords)])

def hessian2(f,x,y):
    return[[f.diff(x).diff(x),f.diff(x).diff(y)],[f.diff(y).diff(x),f.diff(y).diff(y)]]

def typeCrit2(hessian,critPoints):
    '''
    find type of each crit point given

    input: hessian matrix, critical points
    output: [min, max, saddle, unknown, imaginary]
    '''
    
    types = []
    for point in critPoints:
        pointHessian = [[d.subs(x,point[0]).subs(y,point[1]) for d in row] for row in hessian]
        det = pointHessian[0][0]*pointHessian[1][1]-pointHessian[0][1]*pointHessian[1][0]
        if str(det).find("I") != -1:
            types.append("imaginary")
        elif det > 0:            
            if pointHessian[0][0] > 0 or pointHessian[1][1] > 0:
                types.append("min")
            elif pointHessian[0][0] < 0 or pointHessian[1][1] < 0:
                types.append("max")
        elif det < 0:
            types.append("saddle")
        else:
            types.append("unknown")
    return(types)
```

```python
def describeCrit3(f,x,y,z):
    '''
    find location and type of all critical points
    for a function f(x,y,z)

    input: function
    output: [(critical point, type), (critical point, type), ...]
    '''
    # find critical points
    critPoints = critCoords3(f,x,y)
    
    # find type of critical point (2nd erivative testd)
    hessian = hessian3(f,x,y)
    types = typeCrit3(hessian, critPoints)
    
    res = [(critPoints[i],types[i]) for i,_ in enumerate(critPoints)]
    return(res)
    

def critCoords3(f,x,y,z):
    '''
    find coordinates all critical points
    for a function f(x,y)

    input: function
    output: [(x1, y1), (x2, y2), ...]
    '''
    
    # find the partials
    px = f.diff(x)
    py = f.diff(y)
    pz = f.diff(z)
    
    # express y in terms of x
    y_expX = solveset(py,y).args[0]
    
    # substitute to solve for x
    x_eq = px.subs(y,y_expX)
    
    # solve for x coordinates of critical points
    x_coords = [i for i in solveset(x_eq,x).args]
    
    # solve for y coordinates of critical points
    y_coords = [y_expX.subs(x,val) for val in x_coords]
    return([(x_coords[i], y_coords[i]) for i,_ in enumerate(x_coords)])

def hessian2(f,x,y):
    return[[f.diff(x).diff(x),f.diff(x).diff(y)],[f.diff(y).diff(x),f.diff(y).diff(y)]]

def typeCrit2(hessian,critPoints):
    '''
    find type of each crit point given

    input: hessian matrix, critical points
    output: [min, max, saddle, unknown, imaginary]
    '''
    
    types = []
    for point in critPoints:
        pointHessian = [[d.subs(x,point[0]).subs(y,point[1]) for d in row] for row in hessian]
        det = pointHessian[0][0]*pointHessian[1][1]-pointHessian[0][1]*pointHessian[1][0]
        if str(det).find("I") != -1:
            types.append("imaginary")
        elif det > 0:            
            if pointHessian[0][0] > 0 or pointHessian[1][1] > 0:
                types.append("min")
            elif pointHessian[0][0] < 0 or pointHessian[1][1] < 0:
                types.append("max")
        elif det < 0:
            types.append("saddle")
        else:
            types.append("unknown")
    return(types)
```

```python
x = Symbol('x')
y = Symbol('y')
xlamb = np.linspace(-10,10,100)
ylamb = np.linspace(-10,10,100)
X, Y = np.meshgrid(xlamb, ylamb)
```

```python
def func1(x,y):
    return(-x**2+y**2)

Z1 = func1(X, Y)
plt.contour(X,Y,Z1);
f1 = -x**2+y**2
print(describeCrit2(f1,x,y))
```

```python
def hessian(n, f, vs):
    '''
    Returns hessian of a n dimensional function
    as a 2d array
    
    input: (3, f(x,y,z), [x,y,z])
    output: [[fxx, fxy, fxz],[fyx, fyy, fyz],[fzx, fzy, fzz]]
    '''
    
    hessian = []
    
    for row in range(n):
        arow = []
        for col in range(n):
            arow.append(f.diff(vs[row]).diff(vs[col]))
        hessian.append(arow)
    return(hessian)
    
```

```python
f3 = (2*x**2+3*y**2+z**2)*math.e**-(x**2+y**2+z**2)
```
